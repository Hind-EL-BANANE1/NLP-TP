```python
import spacy
```

```python
from spacy.lang.ar import *
```

```python
text="""                                                               .                    """
```

```python
text_ar=Arabic()
text_ar=nlp(text)
```

**# 1 -Tokenization**

```python
for token in text_ar:
    print(token.text)
```

```python
token_list=[]
for token in text_ar :
  token_list.append(token.text)
print(token_list)
```

**# 2 -Pos tagging**

```python
for word in text_ar :
  print(word.text, word.pos_)
```

**Parsing**

```python
from spacy import displacy
displacy.serve(text_ar, style="dep")
```

# **3-Lemmatization**


```python
for word in text_ar:
  print(word.text + ':' ,word.pos_, word.lemma_)
```
