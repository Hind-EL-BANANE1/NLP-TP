```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

**2-Tokenization in spaCy**

```python
text_data=nlp(""" Perhaps one of the most significant advances made by Arabic mathematics began at this time with the work of al-Khwarizmi, namely the beginnings of algebra. It is important to understand just how significant this new idea was. It was a revolutionary move away from the Greek concept of mathematics which was essentially geometry. Algebra was a unifying theory which allowed rational numbers, irrational numbers, geometrical magnitudes, etc., to all be treated as "algebraic objects". It gave mathematics a whole new development path so much broader in concept to that which had existed before, and provided a vehicle for future development of the subject. Another important aspect of the introduction of algebraic ideas was that it allowed mathematics to be applied to itself in a way which had not happened before.""")
```

```python
from spacy.tokenizer import Tokenizer
```

```python
for token in text_data :
  print(token.text)
```

**# Stop words**

```python
spacy_stopword=spacy.lang.en.stop_words.STOP_WORDS
```

```python
print('Number of stop words : %d' % len(spacy_stopword))
```

```python
print('First 10 stop wprds : %s' % list(spacy_stopword)[:20])
```

**# Remove Stop Words**

```python
filtered_sent=[]
for word in text_data:
  if word.is_stop==False:
    filtered_sent.append(word)
print("Filtered Sentence :",filtered_sent)
```

***Recuperer la phrase nominale :***

```python
for chunk in text_data.noun_chunks:
    print(chunk.text, " --> ", chunk.label_)
```

**NER** :
SpaCy dispose d’un système de reconnaissance d’entités statistiques (NER ou Named Entity Recognition) très performant et qui va assigner des étiquettes à des plages contiguës de tokens.

```python
for ent in text_data.ents:
    print(ent.text, ent.label_)
```

**## 3- Part of Speech Tagging**

```python
for word in text_data :
  print(word.text, " : ", word.pos_," : ",word.tag_)
```

*** 4-Lemmatization*** 

```python
for token in text_data:
   print (token.text + ' ==>', token.lemma_)
```

**##5-Word Frequency**


```python
from collections import Counter
```

Remove stop words and punctuation symbols

```python
words = [token.text for token in text_data
        if not token.is_stop and not token.is_punct]
```

```python
word_freq = Counter(words)
```

5 mots courants avec leurs fréquences

```python
common_words = word_freq.most_common(5)
print (common_words)
```
